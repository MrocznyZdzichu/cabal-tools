```python
import CabalTools
main = CabalTools.CabalTools(modules=['DropListManager'])
```

```python
sink = main.DropListManager.preview(
    section_name    = 'World_MobsDrop',
    filter_key      = "WorldName",
    filter_val      = "Forgotten Ruin",
    filter_operator = "like",
)
```

```python
boss_speciesIdx = (108, 109, 110, 111)
for idx in boss_speciesIdx:
    sink = main.DropListManager.preview(
        section_name    = 'World_MobsDrop',
        filter_key      = "SpeciesIdx",
        filter_val      = idx,
        filter_operator = "==",
        suppress        = True
    )
    drop_indices = [int(x) for x in sink['RowIndex'].values]
    for index in drop_indices:
        main.DropListManager.remove_drop_from_list('Mob', index, do_reinit=False, save_files=False)
main.DropListManager.rebuild_scp_index('World_MobsDrop')
```

```python
main.DropListManager.reinit()
sink = main.DropListManager.preview(
    section_name    = 'World_MobsDrop',
    filter_key      = "WorldName",
    filter_val      = "Forgotten Ruin",
    filter_operator = "like",
)
```

```python
for idx in boss_speciesIdx:
    sole_set_1  = CabalTools.MobDrop(WorldIdx = 8, SpeciesIdx=idx, ItemKind = 33557331, ItemOpt = 1,  DropRate = 6)
    sole_set_2  = CabalTools.MobDrop(WorldIdx = 8, SpeciesIdx=idx, ItemKind = 33557331, ItemOpt = 2,  DropRate = 3)
    sole_set_5  = CabalTools.MobDrop(WorldIdx = 8, SpeciesIdx=idx, ItemKind = 33557331, ItemOpt = 5,  DropRate = 1)
    
    drop_list = CabalTools.DropList([sole_set_1, sole_set_2, sole_set_5])
    drop_list.normalize_droprates()

    for drop in drop_list:
        main.DropListManager.add_drop('Mob', drop, do_reinit=False, save_files=False)
main.DropListManager.rebuild_scp_index('World_MobsDrop')
```
